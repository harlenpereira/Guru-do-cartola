```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/henriquepgomide/caRtola/master/data/01_raw/2023/rodada-14.csv'
rodada_atual = pd.read_csv(url)

rodadas_anteriores = []

for i in range(1, 6):
    url = f'https://raw.githubusercontent.com/henriquepgomide/caRtola/master/data/01_raw/2023/rodada-{i}.csv'
    rodada = pd.read_csv(url)
    rodadas_anteriores.append(rodada)
    
df = pd.concat([rodada_atual] + rodadas_anteriores, ignore_index=True)
```

```python
df.isnull().sum()

df2 = df.fillna(0)
```

```python
# Dicionário de mapeamento das posições
mapeamento_posicoes = {
    1: 'GOL',
    2: 'LAT',
    3: 'ZAG',
    4: 'MEI',
    5: 'ATA',
    6: 'TEC'
}

# Criação da coluna "posicao" com base em "atletas.posicao_id"
df2['posicao'] = df2['atletas.posicao_id'].map(mapeamento_posicoes)
```

```python
# Dicionário de mapeamento do status do jogador
mapeamento_posicoes = {
    2: 'PROVÁVEL',
    3: 'CONTUNDIDO',
    5: 'DÚVIDA',
    6: 'SUSPENSO',
    7: 'NULO'
   
}

# Criação da coluna "posicao" com base em "atletas.posicao_id"
df2['status'] = df2['atletas.status_id'].map(mapeamento_posicoes)
```

```python
#Criando FEATURES

df2['desempenho_ultimas_5'] = df2.groupby(['id'])['atletas.media_num'].transform(lambda x: x.rolling(window=5, min_periods=1).mean())


# Preencher valores ausentes com a média da coluna
df2['desempenho_ultimas_5'].fillna(df2['desempenho_ultimas_5'].mean(), inplace=True)

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# X são as características dos jogadores (ex: idade, posição, time, etc.)
# y são as pontuações dos jogadores nas rodadas anteriores


X = df2[['atletas.status_id', 'atletas.rodada_id', 'atletas.pontos_num', 'atletas.clube_id', 'atletas.media_num', 'atletas.atleta_id',
       'atletas.jogos_num', 'atletas.preco_num',
         'atletas.variacao_num',
       'atletas.posicao_id', 'atletas.minimo_para_valorizar', 'DS', 'FC', 'FD',
       'FF', 'FS', 'G', 'V', 'A', 'CA', 'I', 'DE', 'GS', 'PC', 'PS', 'SG',
       'CV', 'FT', 'DP', 'PP', 'GC', 'atletas.temporada_id', 'id']]
y = df2['desempenho_ultimas_5']
#y_y = y.drop(colunas_deletar_y, axis=1)



# Separar os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Treinar o modelo de regressão linear
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Fazer as previsões para a próxima rodada
pontuacoes_previstas = regressor.predict(X_test)

```

```python
pontuacoes_previstas
```

```python
pontuacoes_previstas_df = pd.DataFrame({'ID': X_test['id'], 'Pontuacao Prevista':pontuacoes_previstas})
```

```python
pontuacoes_previstas_df = pontuacoes_previstas_df.sort_values(by='Pontuacao Prevista', ascending = False)
```

```python
pontuacoes_previstas_df.head(30)
```

```python
# Fundir os DataFrames usando o ID como chave de junção
resultado = pontuacoes_previstas_df.merge(df2[['id', 'atletas.apelido', 'atletas.clube.id.full.name','posicao','atletas.status_id', 'status']], left_on='ID', right_on='id', how='left')

# Selecionar as colunas desejadas
resultado = resultado[['ID', 'Pontuacao Prevista', 'atletas.apelido', 'atletas.clube.id.full.name','posicao','atletas.status_id', 'status']]

# Ordenar o resultado pelo valor da coluna "Pontuacao Prevista" em ordem decrescente
resultado = resultado.sort_values(by='Pontuacao Prevista', ascending=False)

# Imprimir o resultado
print(resultado)
```

```python
# Eliminar linhas com valores iguais na coluna "atletas.nome"
resultado_top = resultado.drop_duplicates(subset=['atletas.apelido'])

# Exibir o DataFrame resultante
resultado_top.head(60)
```
